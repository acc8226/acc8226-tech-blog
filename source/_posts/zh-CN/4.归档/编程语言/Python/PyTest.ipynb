```python
# 数据分析示例
import pandas as pd
import numpy as np

# 创建随机数据
data = np.random.randn(5, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# 显示数据
df

# 计算统计信息
df.describe()
```

```python
lst = [2, 1, 3, 9, 5]

for val in lst:   # 遍历元素
    print(val)

for idx, val in enumerate(lst):   # idx 从 0 开始
    print(idx, val)

print(sorted(lst))

print(lst)
```

```python
import random

# 打印 0 - 1 之间的一个小数
print(random.random())
# 生成首尾区间的一个随机整数
print(random.randint(1, 3))
# 获取列表中的一个随机元素 
print(random.choice([1, 2, 3]))

li = [1,2,3,5,6,7]
random.shuffle(li)
print(li)
```

```python
import re

print(re.match(r'\d+', '1998'))
print(re.match(r'1\d{10}', '13566789900'))
```

```python
import time

# 打印时间戳，从 1970年1月1日（UTC/GMT的午夜）开始所经过的秒数
print(time.time())
# 打印当地时间
print(time.localtime())
# 打印指定格式的时间
print(time.strftime('%Y-%m-%d %H:%M:%S'))
```
